# Surface Change Analysis at Lake Mead Using Landsat 8

This notebook analyzes surface change metrics derived from Landsat 8 imagery for the Lake Mead region. Using precomputed spectral indices (NDVI, NDWI, NDBI) from my timelapse tool, we examine temporal trends in vegetation health and surface water extent.

The objective is to demonstrate how satellite-derived geospatial indicators can support 
decision-making related to water availability, environmental constraints, and energy systems 
that depend on surface water resources, such as hydropower and thermal generation cooling.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Improve plot readability
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.grid"] = True
```

```python
import pandas as pd
import numpy as np

# Load CSV
csv_path = "../data/lakemead_surface_change_metrics.csv"
df = pd.read_csv(csv_path)

# Parse date column
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Force numeric conversion
numeric_cols = [
    "mean_ndvi",
    "vegetation_area_m2",
    "urban_area_m2",
    "water_area_m2"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Drop rows with missing values
df = df.dropna(subset=["date", "mean_ndvi"])

# Sort by time
df = df.sort_values("date")

df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["date"].values, df["mean_ndvi"].values)
plt.xlabel("Date")
plt.ylabel("Mean NDVI")
plt.title("Mean NDVI Over Time  Lake Mead Region")
plt.show()
```

The NDVI time series shows an increase beginning in 2014 before a steady decline beginning in 2020, 
with pronounced seasonal variability throughout. This pattern is consistent with prolonged drought conditions 
and reduced soil moisture availability in the Lake Mead watershed.

Vegetation health is a critical environmental indicator for energy system planning, as declining 
vegetation often correlates with increasing land degradation, dust generation, and reduced ecosystem 
resilience.

```python
# Convert square meters to square kilometers for readability
df["water_area_km2"] = df["water_area_m2"] / 1e6

df.plot(x="date", y="water_area_km2", kind="line",
        xlabel="Date", ylabel="Water Surface Area (km)",
        title="Surface Water Area Change  Lake Mead",
        figsize=(10,5), grid=True)
plt.show()

```

The surface water area time series reveals a clear long-term decline, punctuated by short-term 
recoveries likely associated with episodic inflow events. Overall, the downward trend reflects 
well-documented reductions in Lake Mead reservoir levels over the past decade.

Surface water availability is a key constraint for hydropower generation and thermal power plant 
cooling systems. Persistent reductions in reservoir surface area can reduce generation capacity, 
increase operational risk, and influence long-term planning for energy infrastructure in the 
Southwestern United States.


While limited in scale, this analysis demonstrates how satellite-derived geospatial indicators can be translated into 
quantitative, policy-relevant metrics for energy and environmental decision support. By integrating 
spectral indices with area-based calculations, the workflow moves beyond visualization to provide 
measurable indicators of environmental change.
